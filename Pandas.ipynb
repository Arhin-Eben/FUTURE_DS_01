**MAKING CHANGES TO THE DATA**

```python
df.head(5)
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
df['Semi_total'] = df['HP'] + df['Attack'] + df['Defense']
df.head(5)
```

```python
#Dropping a column
df = df.drop(columns=['Semi_total'])
df.head(5)
```

```python
#Dropping all columns except columns 0 to 9
df = df.iloc[:, 0:10]
df.head(5)
```

```python
#exporting the modified to csv
df.to_csv('modified.csv', index=False)
```
